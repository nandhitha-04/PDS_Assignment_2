#a) Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.

```python
import pandas as pd
import numpy as np
#loading the dataset
df = pd.read_csv("/content/train.csv")
df
```

```python
df.info()
```

```python
missing_cols = ['Mileage', 'Engine', 'Power', 'Seats']

# Filling missing values
for col in missing_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())

# Dropping 'New_Price' due to excessive missing data
if 'New_Price' in df.columns:
    df = df.drop(columns=['New_Price'])

print("Missing values handled successfully.")
```

```python
print("Missing values:", df.isnull().sum())
```

```python
from google.colab import files
df.to_csv('data_cleaned_train_A.csv', index=False)
files.download('data_cleaned_train_A.csv')
```

#b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

#C) Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
df2 = pd.read_csv("/content/data_cleaned_train_B.csv")
df2
```

```python
# converting values to 0 and 1
df2['Transmission'] = df2['Transmission'].str.strip()
df2['Fuel_Type'] = df2['Fuel_Type'].str.strip()

df2['Transmission'] = df2['Transmission'].replace({'Automatic': 0, 'Manual': 1})
df2['Fuel_Type'] = df2['Fuel_Type'].replace({'Petrol': 0, 'Diesel': 1});
```

#e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset.

```python
df4 = pd.read_csv("/content/data_cleaned_train_D.csv")
df4
```

```python
# Selecting important columns
selected = df4[['Name', 'Year', 'Fuel_Type', 'Transmission', 'Price']]

# Filtering cars newer than 2015
filtered = df4[df4['Year'] > 2015]

# Renaming columns
renamed = df4.rename(columns={'Mileage': 'Fuel_Efficiency', 'Power': 'Engine_Power'})

# Adding new column for car age
df4['Age'] = 2024 - df4['Year']  # Replace with datetime.now().year if needed

# Sorting by price (highest first)
sorted_by_price = df4.sort_values(by='Price', ascending=False)

# Group by Fuel_Type to get mean price
fuel_mean = df4.groupby('Fuel_Type')['Price'].mean()

# Group by Fuel_Type to get mean, min, max
fuel_stats = df4.groupby('Fuel_Type').agg({'Price': ['mean', 'min', 'max']})

# Print outputs
print("Selected Columns:\n", selected.head())
print("\nFiltered Data (Year > 2015):\n", filtered.head())
print("\nRenamed Columns:\n", renamed.head())
print("\nNew Column 'Age':\n", df4[['Name', 'Year', 'Age']].head())
print("\nSorted by Price:\n", sorted_by_price.head())
print("\nSummary Statistics (Mean Price by Fuel Type):\n", fuel_mean)
print("\nSummary Statistics (Complex):\n", fuel_stats)
```

```python
import os
os.makedirs('results', exist_ok=True)
sections = {
    "Selected Columns:": selected.head(),
    "Filtered Data (Year > 2015):": filtered.head(),
    "Renamed Columns:": renamed.head(),
    "New Column 'Age':": df4[['Name', 'Year', 'Age']].head(),
    "Sorted by Price:": sorted_by_price.head(),
    "Summary Statistics (Mean Price by Fuel Type):": fuel_mean,
    "Summary Statistics (Complex):": fuel_stats
}

# Writing each section to the file
with open('results/results.txt', 'w') as f:
    for title, data in sections.items():
        f.write(f"{title}\n")
        f.write(data.to_string())
        f.write("\n\n")
```
